```python
# dependencies
from bs4 import BeautifulSoup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import time
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

## Scrapping Latest News Title and Paragraph

```python
# Access webpage for data scrapping
browser.visit('https://redplanetscience.com/')
time.sleep(3)
```

```python
# Convert browser html to html string
html = browser.html
```

```python
# Create BeautifulSoup Object and parse with HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Display Soup object
print(soup.prettify())
```

```python
# Retrieve latest news title and news paragraph
news_title = soup.find('div', class_='content_title').text
news_p = soup.find('div', class_='article_teaser_body').text

print(news_title)
print(news_p)
```
