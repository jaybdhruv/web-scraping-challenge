```python
# dependencies
from bs4 import BeautifulSoup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import time
import pandas as pd
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

## Scrapping Latest News Title and Paragraph

```python
# Access webpage for data scrapping
browser.visit('https://redplanetscience.com/')
time.sleep(3)
```

```python
# Convert browser html to html string
news_html = browser.html
```

```python
# Create BeautifulSoup Object and parse with HTML
news_soup = BeautifulSoup(news_html, 'html.parser')
```

```python
# Display Soup object
print(news_soup.prettify())
```

```python
# Retrieve latest news title and news paragraph
news_title = news_soup.find('div', class_='content_title').text
news_p = news_soup.find('div', class_='article_teaser_body').text

print(news_title)
print(news_p)
```

## Scrapping Featured Image

```python
# Access webpage for data scrapping
space_image_url = 'https://spaceimages-mars.com/'
browser.visit(space_image_url)
time.sleep(3)
```

```python
# Using splinter to click on Full Image Button
browser.links.find_by_partial_text('FULL IMAGE').click()
```

```python
# Convert browser html to html string
image_html = browser.html
```

```python
# Create BeautifulSoup Object and parse with HTML
image_soup = BeautifulSoup(image_html, 'html.parser')
```

```python
# Display Soup object
print(image_soup.prettify())
```

```python
# Retrieve Featured Image's Url
featured_image = image_soup.find('img', class_='fancybox-image')['src']
featured_image
```

```python
featured_image_url = space_image_url + featured_image
featured_image_url
```

## Scrapping Mars Fact

```python
# Access webpage for data scrapping
browser.visit('https://galaxyfacts-mars.com/')
time.sleep(3)
```

```python
# Convert browser html to html string
fact_html = browser.html
```

```python
# Create BeautifulSoup Object and parse with HTML
fact_soup = BeautifulSoup(fact_html, 'html.parser')
```

```python
# Display Soup object
print(fact_soup.prettify())
```

```python
# Retrieve comparison table and storing it into a list
comp_table = fact_soup.find('table', class_='table')
comparisons = comp_table.find_all('tr')

mars_fact = []
earth_fact = []
desc = []

for comparison in comparisons:
    
    desc.append(comparison.find('th').text)
    mars_fact.append(comparison.find('span', class_='orange').text)
    earth_fact.append(comparison.find('span', class_='purple').text)

print(desc)
print(mars_fact)
print(earth_fact)
```

```python
# Remove tab part from the list item
earth_fact[-1] = "-88 to 58°C"
```

```python
# Generate Dataframe from the list and set Description column as index
mars_df = pd.DataFrame({"Description":desc, "Mars": mars_fact, "Earth": earth_fact})
mars_df.set_index("Description", inplace=True)
mars_df
```

```python
# Convert dataframe to html
mars_df.to_html('mars_fact.html', classes=["table", "table-striped"], justify="left")
```
